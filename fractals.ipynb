```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import math

import cmath

def complexify(x, y, xmax, ymax, amin, amax, bmin, bmax):
    x_prop = x / xmax
    y_prop = y / ymax
    a_scale = amax - amin
    b_scale = bmax - bmin
    a = x_prop * a_scale + amin
    b = y_prop * b_scale + bmin
    return complex(a,b)

def compute_value(z,f,T,nmax):
    z_n = z
    n = 0
    while (n < nmax and abs(z_n) <= T):
        z_n = f(z_n)
        n += 1
    return n

import matplotlib.colors

cmap = matplotlib.colors.LinearSegmentedColormap.from_list("", ['midnightblue','indigo','darkorange','orangered','rosybrown','floralwhite','white'])
#['darkorange','indigo','midnightblue','orangered','rosybrown','floralwhite','white']

def draw_fractal(f, filename, xmax, ymax, amin, amax, bmin, bmax, T, nmax):
    def combined(x,y): 
        result = compute_value(complexify(x, y, xmax, ymax, amin, amax, bmin, bmax),f,T,nmax)
        return result
    a = np.fromfunction(np.vectorize(combined),(xmax,ymax))
    mpimg.imsave(filename, a, cmap = 'inferno')
    return a


imgs = []
# imgs.append(draw_fractal(lambda z: z**(2.2) - 1.1 + .2j, "extra_1.png", 500, 500, -1,1, -1,1, 10, 10))
# imgs.append(draw_fractal(lambda z: z**3 - .6 + .2j, "extra_2.png", 500, 500, -1,1, -1,1, 100, 10))
# imgs.append(draw_fractal(lambda z:  math.cos(z.imag*2*math.pi)*z**2 - .731 + .3j, "extra_3.png", 500, 500, -1, 1, -1, 1, 1, 4))
# imgs.append(draw_fractal(lambda z: z**(z.real) - z**(z.imag), "extra_4.png", 500, 500, -0.5, 0.5, -0.5, 0.5, 1.3, 10))
# imgs.append(draw_fractal(lambda z: z**(math.sin(z.imag)**2 - z) + math.sin(z.real)*.529 -1.491j, "extra_3.png", 2000, 2000, -1, 1, -1, 1, 9, 20))
# imgs.append(draw_fractal(lambda z: z**(z.real**3 - 1.05*z.real**2 +0.4*z.imag - 0.05*z.real**2) + .48 -1.491j, "extra_2.png", 1500, 1500, -1, 1, -1, 1, 10, 20))
# imgs.append(draw_fractal(lambda z: z**(math.sin(z.real)**2 - z) + .528 -1.491j, "extra_3.png", 1500, 1500, -1, 1, -1, 1, 9.5, 20))
# imgs.append(draw_fractal(lambda z: z**(math.cos(z.real)**2 - 1.3*z) + .565 -1.691j, "extra_7.png", 5000, 5000, -1, 1, -1, 1, 8.9, 20))
# draw_fractal(lambda z: math.tan(z.imag**3 + z.real*z.imag - z.real**2)*(z**(2.2) - z.real*z.imag) - .731 + .3j, "extra_2.png", 1500, 1500, -1, 1, -1, 1, 200, 20)
# draw_fractal(lambda z: math.tan(z.real**3*math.pi + 0.2*z.imag)*(z**3 + z.real*z.imag - z.imag**2) - .731 + .3j, "extra_4.png", 1500, 1500, -1, 1, -1, 1, 500, 20)
#srah: draw_fractal(lambda z: math.cos(z.imag**2*math.e + z.imag**2*math.sin(z.real)*math.pi)*(z**3 + z*z.imag**2 - 0.3*z.real**3 + z) - .731 + .5j, "extra_2.png", 2000, 2000, -1, 1, -1, 1, 1000, 20)


#a = draw_fractal(lambda z: math.cos(z.imag**2*math.e + z.imag**2*math.sin(z.real)*math.pi)*(z**3 + z*z.imag**2 - 0.3*z.real**3 + z) - .731 + .5j, "extra_2.png", 2000, 2000, -1, 1, -1, 1, 1000, 20)
#b = draw_fractal(lambda z: math.sin(z.imag**4*4*math.pi + z.real**3*3*math.pi + z.imag**2*z.imag**3*math.pi*0.5)*(z**3 - 1.5*z.imag*z.real + math.sin(z.real**2 + z.imag**2)) - .931 + .56j, "extra_3.png", 1500, 1500, -1, 1, -1, 1, 500, 20)
#c = draw_fractal(lambda z: math.cos(z.imag**2*4*math.pi + z.real*3*math.pi)*(z**2*z.real - z*z.imag - z.real*z.imag) - .651 + .6j, "extra_4.png", 4000, 4000, -1, 1, -1, 1, 550, 20)
#d = draw_fractal(lambda z: math.cos(z.real*2.5*math.pi - z.imag**2*math.pi)*(z**3*z.imag/(z.real**2 + 0.01) + z**2*z.imag/(z.real**4 + 0.02) - z) - .631 + .5j, "extra_2.png", 4000, 4000, -1, 1, -1, 1, 700, 20)
a = imgs.append(draw_fractal(lambda z: z**(math.sin(z.real)**2 - z) + .528 -1.491j, "extra_3.png", 5000, 5000, -1, 1, -1, 1, 9.5, 20))



# imgs.append(a)
# imgs.append(b)
# imgs.append(c)
# d = (1+2*b)*((a+b)/(b**2 + 0.01*a)) + b*a
# imgs.append(d)
# mpimg.imsave("extra_5.png", d, cmap = 'gist_rainbow')

fig=plt.figure(figsize=(8, 8))
i = 1
for im in imgs:
    fig.add_subplot(3, 2, i)
    plt.imshow(im)
    i += 1
plt.show()
```
